```python
# 테스트 세트를 2차원 배열로 바꿉니다.
test_input = test_input.reshape(-1,1)
print(test_input)
```

```python
# K-최근접 이웃 회귀모델
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor(n_neighbors=5)
knr.fit(train_input, train_target)   #훈련
```

```python
knr.score(train_input, train_target)
# n_neighbors = 1 = 0.9852830341215901
# 3 = 0.9804899950518966
# 5 = 0.9698823289099254
# 7 = 0.9761170732051527
```

```python
# 수동 예측

t_input = test_input[2]
p = knr.predict(t_input.reshape(1,-1))

# 27.5(t_input)에 맞는 weight가 나옴(predict) 
print(t_input.reshape(1,-1))
print(t_input)
print(f"예측 사용 무게: {p}")
print(f"실 무게:{test_target[2]}")
```

```python
# 그래프로 그려보기

knr = KNeighborsRegressor()
x = np.arange(5, 45).reshape(-1, 1)

for k in [1, 3, 5, 7, 9, 10, 15, 21]:
    knr.n_neighbors = k
    knr.fit(train_input, train_target)
    prediction = knr.predict(x)

    # 결정계수 R^2값 추가
    s_train = knr.score(train_input, train_target)
    s_test = knr.score(test_input, test_target)

    plt.title(f'k={k}, R^2={s_train-s_test}')
    plt.scatter(train_input, train_target)
    plt.plot(x, prediction)
    plt.xlabel('length')
    plt.ylabel('weight')
    plt.show()

#R^2 = 차이가 0.00~몇 난다, -값은 좋지 않다. R^2값은 0에 가까울수록 좋다
# k = 3이 가장 좋다.
```

```python
knr = KNeighborsRegressor()
knr.n_neighbors = 42

knr.fit(train_input, train_target)
print(knr.predict([[42]]))    # 42이상보다 값이 더 큰 물고기는 이웃분류기로 불가능
                                # [43]여기서 부터 예측값이 같음 - 예측 범위를 갖고 있음
print(knr.predict([[45]]))
print(knr.predict([[46]]))
print(knr.predict([[47]]))
print(knr.predict([[48]]))
print(knr.predict([[49]]))
print(knr.predict([[50]]))    # 다른 알고리즘 필요
print(knr.predict([[100]]))
```
